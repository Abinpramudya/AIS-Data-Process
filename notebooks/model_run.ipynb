```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import joblib
import os
import csv
```

```python
current_dir = os.getcwd()
root_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
file_name = ''

#input directory
input_directory = os.path.join(root_dir, "data", "DATA AIS NON-STATE")

#model file
model_file = os.path.join(root_dir,"model", "prediction_model.pkl")

#output file
output_path = os.path.join(root_dir,"output", file_name)
```

```python
# Load the saved model
model = joblib.load(model_file)
predictions=''
confidence_level = 0
```

```python
def predict(lat, lon, sog, heading, cog):
    global predictions, confidence_level
    # Load new data for prediction
    new_data = pd.DataFrame({
        'lat [deg]': [lat],
        'lon [deg]': [lon],
        'sog [kn]': [sog],
        'heading [deg]' : [heading],
        'cog [deg]' : [cog]
    })
    # Make predictions on the new data
    predictions = model.predict(new_data)
```

### csv header fixer

```python
def modify_csv_headers(directory, desired_header):
    # Process each file in the directory and its subdirectories
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if filename.endswith(".csv"):
                file_path = os.path.join(root, filename)

                with open(file_path, 'r') as csvfile:
                    reader = csv.reader(csvfile, delimiter=';')  # Specify semicolon delimiter for reading
                    rows = list(reader)

                # Modify only the header row, preserving original content in other rows
                rows[0] = desired_header

                with open(file_path, 'w', newline='') as csvfile:
                    writer = csv.writer(csvfile, delimiter=';')
                    writer.writerows(rows)

                print(f"CSV headers modified successfully for: {file_path}")

# Specify the directory containing the CSV files
directory = os.path.join(root_dir, "output", "predictions")  # Replace with your actual directory

# Desired header
desired_header = [
    "timestamp",
    "lat [deg]",
    "lon [deg]",
    "sog [kn]",
    "draught [m]",
    "heading [deg]",
    "cog [deg]",
    "rot [deg/min]",
    "zone [UTM]",
    "band [UTM]",
    "easting [UTM]",
    "northing [UTM]",
    "STATE",
]
# Call the function to modify CSV headers
modify_csv_headers(directory, desired_header)
```
