```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import joblib
import os
import csv
```

```python
current_dir = os.getcwd()
root_dir = os.path.abspath(os.path.join(current_dir, os.pardir))
file_name = ''

#input directory
input_directory = os.path.join(root_dir, "data", "DATA AIS NON-STATE")

#model file
model_file = os.path.join(root_dir,"model", "prediction_model.pkl")

#output file
output_path = os.path.join(root_dir,"output", file_name)
```

```python
# Load the saved model
model = joblib.load(model_file)
predictions=''
confidence_level = 0
```

```python
def predict(lat, lon, sog, heading, cog):
    global predictions, confidence_level
    # Load new data for prediction
    new_data = pd.DataFrame({
        'lat [deg]': [lat],
        'lon [deg]': [lon],
        'sog [kn]': [sog],
        'heading [deg]' : [heading],
        'cog [deg]' : [cog]
    })
    # Make predictions on the new data
    predictions = model.predict(new_data)
```

```python
# Get a list of all CSV files in the directory
csv_files = [f for f in os.listdir(input_directory) if f.endswith('.csv')]

# Count the number of CSV files
num_csv_files = len(csv_files)

# Print the count
print("Number of CSV files in the folder:", num_csv_files)
state_header = ["STATE"]

for csv_file in csv_files:
    file_name = csv_file
    file_path = os.path.join(input_directory, csv_file)
    output_path = os.path.join(root_dir, "output", "predictions", file_name)

    with open(file_path, 'r') as csvfile:
        reader = csv.reader(csvfile)
        existing_header = next(reader)
        updated_header = existing_header + state_header

    # Create a new CSV file with the updated header
    with open(output_path, 'w', newline='') as csvfile:  # Use 'w' mode to create a new file
        writer = csv.writer(csvfile, delimiter=';')
        writer.writerow(updated_header)  # Write the updated header to the new file

        # Process each row of the existing CSV file and make predictions
        with open(file_path, 'r') as original_csvfile:  # Reopen the original file for reading
            reader = csv.reader(original_csvfile , delimiter=';')
            next(reader)  # Skip the header in the original file
            for row in reader:
                lat = row[1]
                lon = row[2]
                sog = row[3]
                heading = row[5]
                cog = row[6]

                # Call the predict function\
                try:
                    prediction = predict(lat, lon, sog, heading, cog)

                    # Update the row with the prediction
                    row.append(predictions)

                    # Write the updated row to the new CSV file
                    writer.writerow(row)
                except:
                    print('failed on:', csvfile)
                    pass

print('finito')
```
